```python
labels= ['United States', 'China', 'India', 'United Kingdom']

colors=['blue', 'yellow', 'green', 'red' ]

sizes= [423, 165, 41, 33]

plt.pie(sizes,labels=labels, colors=colors, startangle=190, autopct='%1.1f%%')
plt.title('Country wise startup count')

plt.axis('equal')

plt.show()
```
